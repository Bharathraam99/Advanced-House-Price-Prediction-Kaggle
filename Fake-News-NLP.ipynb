```python
data1=pd.read_csv("/kaggle/input/textdb3/fake_or_real_news.csv")
```

```python
data1.head()
```

```python
data2=data1.copy()
```

```python
data2.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
data2['label']=lb.fit_transform(data2['label'])
```

```python
data2.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

cv=CountVectorizer(max_features=5000)
tf=TfidfVectorizer()
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
final=[]
temp=[]
full=[]
for i in range(0,len(data2)):
    tt=data2['text'][i].lower()
    tt=re.sub('[^a-zA-Z]'," ",tt)
    temp=tt.split()
    final=[lemmatizer.lemmatize(i) for i in temp if not i in stopwords.words('english')]
    full.append(" ".join(final))
    
    
```

```python
from sklearn.naive_bayes import MultinomialNB
mb=MultinomialNB()
```

```python
X=cv.fit_transform(full).toarray()
```

```python
y=data2['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
mb.fit(X_train,y_train)
```

```python
y_pred=mb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```
